```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
def L(p,q):
    return ((1-p)*np.sqrt(q/(1-q)))+(p*np.sqrt((1-q)/q))-((1-p)*np.sqrt(p/(1-p)))-(p*np.sqrt((1-p)/p))
```

```python
p = np.linspace(0.1,0.9,30)
q = np.linspace(0.1,0.9,30)
X,Y = np.meshgrid(p,q)
Z = L(X,Y)
```

```python
figure = plt.figure()
ax = plt.axes(projection = '3d')
ax.contour3D(X, Y, Z, 50, cmap ='binary')
ax.set_xlabel('p')
ax.set_ylabel('q')
ax.set_zlabel('L')
```

```python
def Lc(p,q):
    return ((1-p)*np.sqrt(q))+(p*np.sqrt(1-q))
```

```python
p = np.linspace(0.1,0.9,30)
q = np.linspace(0.1,0.9,30)
X,Y = np.meshgrid(p,q)
Z = Lc(X,Y)
```

```python
figure2 = plt.figure()
ax = plt.axes(projection = '3d')
ax.contour3D(X, Y, Z, 50, cmap ='binary')
ax.set_xlabel('p')
ax.set_ylabel('q')
ax.set_zlabel('L')
```
